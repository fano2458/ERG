```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load the CSV file
csv_file_path = r"C:\Users\haise\Desktop\erg\conrangedataset\02.20-03.20.csv"  # replace with your CSV file path
df = pd.read_csv(csv_file_path)

# Specify the column name you're looking for
column_name = '1023-115-201-GAZO-0001-GCP1/G_BB02_PIA03/FurOverpressure.U'

# Assuming there is a datetime column named 'time' or similar in your CSV
datetime_column = 'time'  # replace with the actual datetime column name in your CSV

# Convert the datetime column to datetime if it's not already
df[datetime_column] = pd.to_datetime(df[datetime_column], errors='coerce')

# Check if the column exists in the DataFrame
if column_name in df.columns:
    # Get all unique dates from the datetime column
    unique_dates = df[datetime_column].dt.date.unique()
    
    # Create a directory to save the plots
    output_dir = 'plots'
    os.makedirs(output_dir, exist_ok=True)
    
    # Iterate over each unique date and plot the data
    for date in unique_dates:
        filtered_df = df[df[datetime_column].dt.date == date]
        
        plt.figure(figsize=(10, 5))
        plt.plot(filtered_df[datetime_column], filtered_df[column_name], marker='o', linestyle='-')
        plt.title(f'Values of {column_name} on {date}')
        plt.xlabel('Date and Time')
        plt.ylabel('Values')
        plt.grid(True)
        
        # Save the plot as a PNG file
        plot_file_path = os.path.join(output_dir, f'{date}.png')
        plt.savefig(plot_file_path)
        plt.close()
    
    print(f'Plots saved in directory: {output_dir}')
else:
    print(f'Column "{column_name}" not found in the CSV file.')
```
